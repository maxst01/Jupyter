```python
from sympy import *
from sympy.plotting import plot

x = Symbol('x')
func = -12*x**4*sin(cos(x))-18*x**3+5*x**2+10*x-30
plot(func,(x,-8, -7.5),line_color='red')
```

```python
roots = []
root = nsolve(func, -7.65)
roots.append(root)
display(root)
```
